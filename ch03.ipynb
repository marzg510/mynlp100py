### 20. JSONデータの読み込み

```python
import json

article_json = []
with open('jawiki-country.json') as f:
  line = f.readline()
  while line:
    article_json.append(json.JSONDecoder().raw_decode(line))
    line = f.readline()
#  print(f.readlines())
#  article_json = json.loads(",".join(f.readlines()))

for i,a in enumerate(article_json):
  print(i,a[0]["title"])
```


### 21. カテゴリ名を含む行を抽出

```python
#import re
#m = re.match(r'Category:', article_json[0][0]['text'])
for i,a in enumerate(article_json):
  if "[[Category:" in article_json[0][0]['text']:
    print(i)

```

### 22. カテゴリ名の抽出

```python
import re
categories = set()
for i,a in enumerate(article_json):
  m = re.findall(r'\[\[Category:(.*)\]\]', a[0]['text'])
  categories |= set(m)
print(categories)
```

### 23. セクション構造

```python
sections = set()
for a in article_json:
  ms = re.findall(r'==+ .+ =+=',a[0]["text"])
  for m in ms:
    name = re.search(r'=*= (.+) =+=',m)
    lvl = int(m.count('=') / 2) - 1
#    sections.add({"name":name, "level":lvl})
    sections.add("{} {}".format(name.groups()[0],lvl))

sections
```


### 24. ファイル参照の抽出


### 25. テンプレートの抽出

```python
#article_json[0][0]["text"]
```

```python

#re.findall(r"\{\{基礎情報.*\}\}",article_json[0][0]["text"])
#b = re.findall(r"{{基礎情報(.*)}}",''.join(article_json[0][0]["text"].splitlines()))
#print(re.findall(r".*}}",article_json[0][0]["text"]))
#print(article_json[0][0]["text"])
#b
bases = []
for i,a in enumerate(article_json):
  if i > 0:
    break
  is_base = False
  base = {}
  for l in a[0]["text"].splitlines():
    if "{{基礎情報 " in l:
      is_base = True 
      continue
    if ("}}" == l) and is_base:
      is_base = False
      continue
    if is_base:
      m = re.search(r"\|([^=]+)=(.*)$",l)
      base[m.groups()[0].strip()] = m.groups()[1]
  bases.append(base)
  
bases


```


### 26. 強調マークアップの除去

```python
base2 = {}
for k,v in list(base.items()):
  v = re.sub(r"'''''(.*)'''''",r"\1",v)
  v = re.sub(r"'''(.*)'''",r"\1",v)
  v = re.sub(r"''(.*)''",r"\1",v)
  base2[k] = v
#''他との区別''
#'''強調'''
#'''''斜体と強調'''''
base2
```


### 27. 内部リンクの除去

```python
base3 = {}
for k,v in list(base2.items()):
  v = re.sub(r"(?<=\[\[)(.+?)(?=\]\])",r"\1",v)
  m = re.findall(r"(?<=\[\[).+?(?=\]\])",v)
  base3[k] = ''.join(m) if m else v
base3
```


### 28. MediaWikiマークアップの除去

```python
base4 = {}
for k,v in list(base3.items()):
  v = re.sub(r"file:",r"",v)
  v = re.sub(r"File:",r"",v)
  v = re.sub(r"ファイル:",r"",v)
  base4[k] = v
base4
```


### 29. 国旗画像のURLを取得する

```python
import urllib

url = 'https://www.mediawiki.org/w/api.php?' \
    + 'action=query' \
    + '&titles=File:' + urllib.parse.quote(base4["国旗画像"]) \
    + '&format=json' \
    + '&prop=imageinfo' \
    + '&iiprop=url'
print(url)
request = urllib.request.Request(url)
#    headers={'User-Agent': 'NLP100_Python(@segavvy)'})
connection = urllib.request.urlopen(request)
# jsonとして受信
data = json.loads(connection.read().decode())
#file_url = data['query']['pages'].popitem()[1]['imageinfo'][0]['url']
#print("file url:{}".format(file_url))
```
